```python
# EJECUTATE
# Recorro todas las páginas con el shortcut y recojo los links en la lista links_dinos

short_cut = driver.find_elements_by_class_name("category-page__alphabet-shortcut")

links_dinos = []
lista_nombres = []

for i in range(1,len(short_cut)):
    print(i)
    
    dinos = driver.find_elements_by_class_name("category-page__member")
    
    for dino in dinos:
        link = dino.find_element_by_tag_name('a').get_attribute('href')
        print(link)
        links_dinos.append(link)
        
        nombre_dino= dino.find_element_by_tag_name('a').get_attribute('title')
        print(nombre_dino)
        lista_nombres.append(nombre_dino)
        
    short_cut = driver.find_elements_by_class_name("category-page__alphabet-shortcut")
    
    try:
        short_cut[i].click()
    
    except:
        print(short_cut[i].text,"no se puede clicar")

    #print(lista_nombres)
    #print(links_dinos)
```

```python
# EJECUTATE
# Voy a dividirlo en 5 partes
link_dinos1 = df_links["links"].iloc[0:200] 
link_dinos2 = df_links["links"].iloc[200:400]
link_dinos3 = df_links["links"].iloc[400:600]
link_dinos4 = df_links["links"].iloc[600:800]
link_dinos5 = df_links["links"].iloc[800:]

# Observo último de cada parte
print(link_dinos1[-1:])
print(link_dinos2[-1:])
print(link_dinos3[-1:])
print(link_dinos4[-1:])
print(link_dinos5[-1:])
```

```python
# EJECUTATE
# Mi función que me extrae los datos que necesito

def extraer_info_dinos():
    '''
    Función que extrae de cada página el elemento que contiene los datos que quiero buscar, y me extraigo los que me interesan
    Outputs - 3 -
        nombre: string
        altura: string
        peso: string
    '''
    try: 
        caract = driver.find_elements_by_tag_name("aside")
           
        list_cuadro = caract[0].text.split("\n")
        nombre = list_cuadro[0]
        if "Altura" in list_cuadro:
            indice = list_cuadro.index("Altura")
            altura = list_cuadro[indice + 1]
        else: 
            altura = "Nan"
        if "Peso" in list_cuadro:
            indice2 = list_cuadro.index("Peso")
            peso = list_cuadro[indice2 + 1]
        else:
            peso = "Nan"
    except:
        nombre = "sin_info"
        altura = "sin_info"
        peso = "sin_info"

    return nombre, altura, peso
  
```

```python
# PRUEBAS
# Compruebo varios dinosaurios, y me saca en las listas la información que necesito
list_nombres = []
list_alturas = []
list_pesos = []
nombre_dino, altura_dino, peso_dino = extraer_info_dinos()
list_nombres.append(nombre_dino)
list_alturas.append(altura_dino)
list_pesos.append(peso_dino)

print(list_nombres)
print(list_alturas)
print(list_pesos)
```

```python
# EJECUTATE
# Función para crear las listas por partes 

def crear_listas(links):
    '''
    Función que me almacena en listas los 3 tipos de datos que se han extraido
    Input - 1 - 
        lista de links
    '''
    for link in links:
        driver.execute_script("window.open("");")
        driver.switch_to.window(driver.window_handles[-1])
        driver.get(link)
    
        nombre_dino, altura_dino, peso_dino = extraer_info_dinos()
        list_nombres.append(nombre_dino)
        list_alturas.append(altura_dino)
        list_pesos.append(peso_dino)
    
        driver.close()
        driver.switch_to.window(driver.window_handles[-1])        
```
