## Extraemos los links de la página individual para cada dinosaurio

```python
contenedor = driver.find_elements_by_class_name("fwpl-item.el-orhxu")
```

```python
# CREAR LISTA LINKS: EJECUTATE

lista_links_abc = []

for item in contenedor:
    link = item.find_element_by_tag_name('a').get_attribute('href')
    print(link)
    try:
        lista_links_abc.append(link)
    except:
        print("El link",link,"no se añade")
        

lista_links_abc
```

```python
# EJECUTATE
# Guardo los links en un dataframe y este en un csv
df_links_abcdinos = pd.DataFrame({"Links ABCDinos": lista_links_abc})

df_links_abcdinos
```

```python
# EJECUTATE
# Dataframe convertido a csv
df_links_abcdinos.to_csv('df_links_abcdinos.csv', index = False)
```

```python
def extraer_info_dinos():
    '''
    Función que extrae de cada página la info que me interesa
    Outputs - 3 -
        nombre: string
        altura: string
        peso: string
    '''
    try: 

        nombre_tablita = driver.find_element_by_class_name('elementor-repeater-item-add638e').text
        altura_tablita = driver.find_element_by_class_name('elementor-repeater-item-40ceb8d').text
        peso_tablita = driver.find_element_by_class_name('elementor-repeater-item-4919f23').text
    except:
        nombre_tablita = "sin_info"
        altura_tablita = "sin_info"
        peso_tablita = "sin_info"

    return nombre_tablita, altura_tablita, peso_tablita
```

```python
# EJECUTATE
# 1ªPARTE: recorro y guardo en listas los contenidos que me interesan

list_nombres = []
list_alturas = []
list_pesos = []

for link in lista_links_abc:
    driver.execute_script("window.open("");")
    driver.switch_to.window(driver.window_handles[-1])
    driver.get(link)
    
    nombre_dino, altura_dino, peso_dino = extraer_info_dinos()
    list_nombres.append(nombre_dino)
    list_alturas.append(altura_dino)
    list_pesos.append(peso_dino)
    
    driver.close()
    driver.switch_to.window(driver.window_handles[-1])
    

print(list_nombres)
print(list_alturas)
print(list_pesos)
```

```python
# EJECUTATE
# Paso a dataframe la info obtenida en las 2 listas
df_dinos_abcdinos = pd.DataFrame({"Nombre": list_nombres, "Altura" : list_alturas,"Peso" : list_pesos})
df_dinos_abcdinos
```

```python
df_dinos_abcdinos.to_csv('df_dinos_abcdinos.csv', index = False)
```
